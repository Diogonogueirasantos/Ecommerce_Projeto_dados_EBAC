```python
df.drop(index=[69, 79, 136, 140, 190, 196, 252, 255], inplace=True)
```

```python
df['sazonalidade'] = df['Temporada'].replace({"outono/inverno":"6 meses menor temperatura",
                                               "primavera/verão":"6 meses maior temperatura",
                                               "não definido":"demanda regular",
                                               "primavera-verão outono-inverno":"alta frequência",
                                               "primavera-verão - outono-inverno":"alta frequência"})
```

```python
df[df['Temporada'] == "2021"]
```

```python
df.head()
```

```python
df.drop(columns=["Nota_MinMax", "N_Avaliações_MinMax", "Desconto_MinMax", "Preço_MinMax",
                 "Temporada_Cod", "Qtd_Vendidos_Cod", "Marca_Freq", "Review1", "Review2", "Review3",
                 "Temporada", "Material_Freq"], inplace=True)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.drop(columns=["Unnamed: 0", "Marca_Cod", "Material_Cod"], inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"Título":"produto", "Nota":"nota", "N_Avaliações":"avaliacoes",
                   "Desconto":"desconto", "Marca":"marca", "Material":"material",
                   "Gênero":"publico_voltado", "Qtd_Vendidos":"qtd_vendidos", "Preço":"preco",}, inplace=True)
```

```python
df.head()
```

```python
df['publico_voltado'].unique()
```

```python
df['publico_voltado'] = df['publico_voltado'].replace({"Sem gênero":"Unissex", "Meninos":"Infantil:masculino",
                                                       "Meninas":"Infantil:feminino", "Sem gênero infantil":"infantil:Unissex"})
```

```python
df['publico_voltado'].unique()
```

```python
df['avaliacoes'] = df['avaliacoes'].astype(int)
```

```python
df['desconto'] = df['desconto'].astype(int)
```

```python
df.head(2)
```

```python
df['qtd_vendidos'].unique()
```

```python
df['qtd_vendido_int'] = df['qtd_vendidos'].replace({"+5":5, "+25":25, "+50":50, "+100":100,
                                                    "+1000":1000,"+10mil":10000, "+50mil":50000})    
```

```python
df['preco_desconto'] =  round(df['preco'] - (df['preco'] * df['desconto'] / 100), 2)
```

```python
df.drop(columns=['qtd_vendidos'], inplace=True)
```

```python
df = df.rename(columns={"qtd_vendido_int":"qtd_vendidos"})
```

```python
df.head()
```

```python
# ordenando variáveis

ordem = ['produto', 'marca', 'material', 'preco', 'desconto', 'preco_desconto', 'publico_voltado', 'nota', 'avaliacoes', 'qtd_vendidos', 'sazonalidade']
```

```python
df = df[ordem]
```

```python
df.head()
```

```python
df.to_csv('ecommerce_estatistica_V1.csv', index=False)
```
